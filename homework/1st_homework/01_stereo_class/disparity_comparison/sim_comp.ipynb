```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read disparity maps
orig1 = cv2.imread("./teddy/im2.png")
orig2 = cv2.imread("./teddy/im6.png")

disp_our = cv2.imread("./teddy_disparity_naive.png")
disp_orig = cv2.imread("./teddy/disp2.png")
```

```python
def show_images(img1, img2, img3, img4, figsize=(3,3)):
    f, axarr = plt.subplots(2,2)
    axarr[0,0].imshow(img1, cmap='gray')
    axarr[0,1].imshow(img2, cmap='gray')
    axarr[1,0].imshow(img3, cmap='gray')
    axarr[1,1].imshow(img4, cmap='gray')

    plt.show()
```

```python
show_images(orig1, orig2, disp_our, disp_orig)

print(disp_orig.shape)
print(disp_our.shape)

```

```python
def ncc(img1, img2):
    img1 = (img1 - np.mean(img1)) / (np.std(img1) * len(img1))
    img2 = (img2 - np.mean(img2)) / (np.std(img2))
    return np.correlate(img1, img2, 'full')  
```

```python
ncc(disp_our, disp_orig)
```
